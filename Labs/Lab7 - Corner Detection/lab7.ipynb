```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('chessBoard.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# turn the image matrix into floats, so that when we apply corner detection, we can get accurate float values.
gray = np.float32(gray)

# 2 is the stride
# kernal size is 3x3
# alpha value is 0.04
corners = cv2.cornerHarris(gray, 2, 3, 0.04)

#result is dilated for marking the corners, not important
kernel = np.ones((7, 7), np.uint8)

corners = cv2.dilate(corners,kernel)

# Threshold for an optimal value, it may vary depending on the image.
# corners > 0.3 * corners.max() is the threshold.
img[corners > 0.3 * corners.max()] = [0, 0, 255]
```
